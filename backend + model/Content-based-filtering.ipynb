```python
data = data.drop(["crawl_timestamp","uniq_id","is_FK_Advantage_product"],axis=1)
```

```python
data = data.dropna()
data = data.drop_duplicates(subset=["product_name"])
```

```python
# Remove '.','>' from all the values
import ast
def cleaner(x):
    literal = ast.literal_eval(x)
    y = "".join(literal)
    return y.replace(">","").replace(",","")
```

```python
data.loc[:,"product_category_tree"] = data["product_category_tree"].apply(cleaner)
```

```python
data.iloc[:]
```

```python
def combine(x):
    return x['product_name'] + ' ' +(x['description'] if pd.notnull(x['description']) else '') + ' '+ x['product_category_tree'] + ' '+ (x['brand'] if pd.notnull(x['brand']) else '')
data['combined'] = data.apply(combine, axis=1)
```

```python
data = data.reset_index(drop=True)
data
```

```python
data
```

```python
# Create an instance of CountVectorizer with stop words in English
count = CountVectorizer(stop_words='english')

# Fit and transform the 'combined' column of the 'filtered' DataFrame using CountVectorizer
count_matrix = count.fit_transform(data['combined'])
```

```python
cosine_sim_countvec = cosine_similarity(count_matrix, count_matrix)  # Calculate Cosine Similarity
cosine_sim_countvec
```

```python
indices = pd.Series(data.index, index=data['pid']) # Mapping title with the index
```

```python
indices
```

```python
def recommend(pid,cosine_sim):
    # Get the index of the given 'title' in the 'indices' list or dictionary
    index = indices[pid]
    
    if(not index):
        return "Item Not Found"
    # Get the cosine similarity of the given title with all other titles
    similarity = list(enumerate(cosine_sim[index]))
    
    # Sort the similarity scores in descending order
    similarity = sorted(similarity, key=lambda x: x[1], reverse=True)
    
    # Get the top 15 similar titles (excluding the given title itself)
    scores = similarity[1:16]
    
    # Get the original indices of the top 15 similar titles
    org_index = [i[0] for i in scores]
    similar_titles = data[['pid','product_name', 'description']].iloc[org_index].copy()
    similar_titles['similarity'] = [i[1] for i in scores]
    # Return a DataFrame with the 'productId', 'name', and 'description' columns of the similar titles
    return similar_titles
```

```python
recommend("ACCEJWHTY2HVMG7Z5",cosine_sim_countvec)
```

```python
# Create an instance of TfidfVectorizer with stop words removed
tfidf = TfidfVectorizer(stop_words='english')

# Compute the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(data['combined'])
```

```python
cosine_sim_tfidf = cosine_similarity(tfidf_matrix,tfidf_matrix) # calculate cosing similarity
cosine_sim_tfidf
```

```python
recommend("ACCEJWHTY2HVMG7Z5",cosine_sim_tfidf) # Recommend using tfidf
```

```python
data.to_csv('filtered.csv',index=False) # Save dataframe in filtered.CSV
pickle.dump(cosine_sim_countvec,open('cosine_sim_countvec.pkl','wb')) # Save Count Vector Cosine Similarity
pickle.dump(cosine_sim_tfidf,open('cosine_sim_tfidf.pkl','wb')) # Save TFIDF Cosine Similarity
```
